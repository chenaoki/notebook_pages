```python
import numpy as np
import matplotlib.pyplot as plt
import os
import pandas as pd
import pickle
import copy
import cv2
import math
from matplotlib import animation, rc
from IPython.display import HTML
%matplotlib inline
```

```python
exp_dir = "/mnt/recordings/ExperimentData/"
anl_dir = "/mnt/recordings/AnalysisResult/Tomii/2017-07-10/"

df_path = anl_dir + "pv_stim.csv"

pv_thre = 0.7
```

```python
df = pd.read_csv(df_path)
print "N:",len(df)
df
```

```python
sessions = range(len(df))

#sessions =  range(15, 25)

#sessions = [14]
```

```python
## 刺激パラメータ

stim_params = np.zeros([len(df), 4])

def phase_comp(p):
    return math.atan2(np.absolute(np.sin(p)), np.absolute(np.cos(p)))

for i in sessions:

    s = df.loc[i,:]
    sample = s["sample"]
    session = s["session"]
    target = "{0}/{1}/C001H001S{2:0>4}".format(exp_dir, sample, session)    
    save_dir = target.replace(exp_dir, anl_dir)
    
    #file_path = os.path.join(save_dir, 'core.log')
    #os.system('cp {0} {1}'.format(file_path, file_path.replace('core', 'core_extracted')))
    #continue

    #vmem = np.load( os.path.join(save_dir, "vmem.npz") )["arr_0"]
    #pmap = np.load( os.path.join(save_dir, "pmap.npz") )["arr_0"]
    #print pmap.shape
    
    a = np.loadtxt(os.path.join(save_dir, "elec_phase.log"), delimiter=' ')
    
    plt.clf()
    plt.plot( a[:,1])
    f_stim = s["frame_stim"]
    
    #plt.plot( [ np.pi, s["frame_stim"]], [ -np.pi, s["frame_stim"]], 'k-' )
    plt.plot( [ f_stim, f_stim], [ np.pi, -np.pi], 'r-', lw = 1)
    plt.ylim( [ np.pi, -np.pi] )
    plt.savefig(os.path.join(save_dir, 'plot_phi.png'))
    
    phi = a[ f_stim-1, 1]
    
    vec_pre = (s.stim_x - s.pre_x) + 1j*(s.stim_y - s.pre_y)
    theta = phase_comp(np.angle(vec_pre) - s.fb_angle)
    l = np.abs(vec_pre) * 1000 / s.speed
    
    # Normalized shift 
    shift = np.abs(((s.pst_x - s.pre_x) + 1j*(s.pst_y - s.pre_y))) * 1000 / s.speed
    
    d = np.array([phi, theta, l, shift])
    stim_params[i,:] = d
    print '{0}:{1}:{2}' .format(i, save_dir, d)
```

```python
shift_thre = 10
step_t1 = 8
step_t2 = 8
step_l = 8
```

```python
plt.hist(stim_params[:,3], np.arange(1,40,2))
plt.title("histogram of sprial shift distance")
plt.show()
print np.mean(stim_params[:,3])
```

```python
data_num = np.zeros((step_t1, step_t2, step_l))
data_suc = np.zeros((step_t1, step_t2, step_l))

success =(stim_params[:,3] > shift_thre) * 1.0
phi = ( ( stim_params[:,0]+np.pi) / ( np.pi / (step_t1/2) ) ).astype(np.int8)
theta = (stim_params[:,1] / ( np.pi / (2*step_t2) ) ).astype(np.int8)
l = (stim_params[:,2] / ((np.max(stim_params[:,2])+1)/step_l)).astype(np.int8)

for n , s in enumerate(success):
    i = phi[n]
    j = theta[n]
    k = l[n]
    
    data_num[i, j, k] += 1
    if s > 0: data_suc[i, j, k] += 1
```
