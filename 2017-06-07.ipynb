```python
%matplotlib inline
import json
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import animation
from IPython.display import HTML

import elecpy.elecpy as elp
from elecpy.util.cmap_bipolar import bipolar
```

```python
elp = reload(elp)
with open ('./elecpy/temp/sim_params_ext.json','r') as f:
    elp.sim_params = json.load(f)
print json.dumps(elp.sim_params, indent=4)
```

```python
im_h   = elp.sim_params['geometory']['height']
im_w  = elp.sim_params['geometory']['width']

fig, ax = plt.subplots()

im = plt.imshow(
    np.zeros((im_h,im_w),dtype=np.float64),
    vmin = -100.0, vmax = 100.0,
    cmap=bipolar(neutral=0, lutsize=1024),
    interpolation='nearest')

plt.close()

g = elp.sim_generator()

def init():
    im.set_array(np.zeros((im_h,im_w),dtype=np.float64))
    return (im,)

def draw(data):
    try:
        vmem = g.next()
        im.set_array(vmem)
        return (im,)
    except StopIteration:
        return init()

plt.rcParams['animation.html'] = 'html5'

anim = animation.FuncAnimation(
    fig, draw, init_func = init, 
    save_count = elp.conv_time2cntSave(elp.sim_params['time']['end']),
    blit=False, interval=500, repeat=False)
anim
```
