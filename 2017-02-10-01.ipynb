```python
import matplotlib.pyplot as plt
import numpy as np
from  scipy.signal import hilbert, chirp
%matplotlib inline
```

```python
dt = 0.01
n = 128
```

```python
t = np.linspace(1, n, n)*dt - dt
x = chirp(t, 2.0, t[-1], 10.0)
x *= (1.0+0.8*np.sin(2.0*np.pi*3.0*t))
y = hilbert(x)
plt.plot(x)
plt.plot(np.imag(y))
plt.show()
```
