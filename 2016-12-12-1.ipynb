```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
T = np.linspace(-100*np.pi,100*np.pi,10000)
```

```python
a = 1.0
b = 0.5
```

```python
r = a*np.exp(b*T)
for i in range(6):
    r*=1.7
    plt.plot(r*np.cos(T), r*np.sin(T), 'b')
t = np.linspace(-np.pi, np.pi, 1000)
plt.plot(10*np.cos(t), 10*np.sin(t), 'k')

plt.xlim([-20,20])
plt.ylim([-20,20])
plt.show()
```
